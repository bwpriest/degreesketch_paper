```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from glob import glob
from os import listdir
```

```python
eval_files = glob('nbhd_eval/*.csv')
evals = [pd.read_csv(f, sep='\s*,\s*') for f in eval_files]
```

```python
evals[0].keys()
```

```python
fig, ax = plt.subplots()

labelsize = 18
ticksize = 14

colors = ["purple", "indigo", "blue", "green", "red",
          "violet", "olive", "orange", "aquamarine", "magenta"]

color = "black"
for ev, clr in zip(evals, colors):
    plt.plot(range(1,6), ev['mre'].to_numpy(), '-', color=clr)
ax.set_xticks(range(1, 6))
ax.set_xticklabels(['1', '2', '3', '4', '5'], fontsize=ticksize)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(ticksize) 
ax.set_ylabel("Mean Relative Error", color=color, fontsize=labelsize)
ax.tick_params(axis='y', labelsize=ticksize, labelcolor=color)
ax.set_xlabel("Pass Number", fontsize=labelsize)

# ax2 = ax.twinx()
# color="darkblue"
# for ev, clr in zip(evals, colors):
#     plt.plot(range(1,6), ev['mean time'].to_numpy(), '--', color=clr)
# ax2.set_ylabel("Mean Wall Time (s)", color=color, fontsize=labelsize)
# ax2.tick_params(axis='y', labelsize=ticksize, labelcolor=color)
# for tick in ax2.yaxis.get_major_ticks():
#     tick.label.set_fontsize(ticksize) 
#     tick.label.set
plt.tight_layout()
fig.savefig("foo.pdf", bbox_inches='tight')
```
